```python
data=data.fillna({'Video Uploads':0,'Subscribers':0})
```

```python
data[['Video Uploads','Subscribers']]=data[['Video Uploads','Subscribers']].astype(int)
```

### 9.Visualization Of Above Data

```python
data.columns
```

```python
Channel_name=['Zee TV','T-Series','Cocomelon','SET India','WWE']
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(60,30))
fig.suptitle("Top-5-Youtube-Channels Dataset",fontsize=55)
ax1.pie(data['Subscribers'].head(),labels=Channel_name,autopct="%0.2f%%",radius=1.7,shadow=True,textprops={'fontsize':40})
ax1.set_title("YouTube's Most Subscribed Channel",y=1.3)

#SET FONTSIZE 1
font_size=45
for item in ([ax1.title,ax1.xaxis.label,ax1.yaxis.label]+ax1.get_xticklabels()+ax1.get_yticklabels()):
    item.set_fontsize(font_size)

ax2.bar(Channel_name,data['Video Uploads'].head(),align="center",y=-1.5)
ax2.set_xlabel("Name Of The Channel Name",labelpad=40)
ax2.set_ylabel("No of Video's upload",labelpad=40)
ax2.set_ylim(0,100000)
ax2.set_title("YouTube's Most Uploaded Videos",y=1.3)

#SET FONTSIZE 2
font_size=45
for item in ([ax2.title,ax2.xaxis.label,ax2.yaxis.label]+ax2.get_xticklabels()+ax2.get_yticklabels()):
    item.set_fontsize(font_size)

ax3.scatter(Channel_name,data['Video views'].head(),s=800,c=['blue','black','violet','red','green'])
ax3.set_title("YouTube's Most Viewed Channel")
ax3.set_xlabel("Name Of The Channel Name",labelpad=50)
ax3.set_ylabel("No of Video views",labelpad=40)


#SET FONTSIZE 3
font_size=45
for item in ([ax3.title,ax3.xaxis.label,ax3.yaxis.label]+ax3.get_xticklabels()+ax3.get_yticklabels()):
    item.set_fontsize(font_size)


ax4.plot(data[ 'Subscribers'].head(5),linewidth=6,marker='>',markersize=30,markerfacecolor='red')
ax4.set_title("Top 5 YouTube's Channel")
ax4.set_xlabel("Name Of The Channel Name",labelpad=40)
ax4.set_ylabel("Top 5 Subscribers",labelpad=40)
ax4.set_xticks([0,1,2,3,4])
ax4.set_xticklabels(Channel_name)
    
#SET FONTSIZE 4
font_size=45
for item in ([ax4.title,ax4.xaxis.label,ax4.yaxis.label]+ax4.get_xticklabels()+ax4.get_yticklabels()):
    item.set_fontsize(font_size)


plt.subplots_adjust(hspace=-1)
plt.tight_layout()
plt.savefig("Top-5-Youtube-Channels Dataset",dpi=17)
plt.show()




```
